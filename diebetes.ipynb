```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# Example for Min-Max Scaling
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

# Initialize PCA and set the number of components
pca = PCA(n_components=2)  # Reduce data to 2 components
principal_components = pca.fit_transform(scaled_df)

# Explained variance ratio
print(pca.explained_variance_ratio_)  # Shows how much variance is captured by components
```

```python
import matplotlib.pyplot as plt

plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization')
plt.show()
```
